### Data cleaning
- Remove Unnecessary Columns
- Data Type Mismatch

```python
import sys
from pathlib import Path
PROJECT_ROOT = Path().resolve().parent
sys.path.append(str(PROJECT_ROOT))
```

```python
# Import Modules
import numpy as np
import pandas as pd
from src import config
```

```python
# Load data 
DATA_FILE_PATH = config.DATA_PATH
CLEANED_DATA_FILE_PATH = config.CLEANED_DATA_PATH
print("Raw Data Path:\n", DATA_FILE_PATH)
print("Cleaned Data Path:\n", CLEANED_DATA_FILE_PATH)
```

```python
#Load data 
data = pd.read_csv(DATA_FILE_PATH)
data.head()
```

```python
# Data back up
df = data.copy()
```

```python
# Meta data
print("No of rows:",df.shape[0])
print("No of columns:",df.shape[1])
print("All columns:\n",df.columns.to_list())
print("Data types:\n",df.dtypes)
print("\nInfo:\n")
df.info()
```

```python
# Change data type of "Total Charges"
df["TotalCharges"]= pd.to_numeric(df["TotalCharges"],errors= "coerce")
print("Data type of 'TotalCharges':",df["TotalCharges"].dtypes)
```

```python
# Remove irrelevant columns
df.drop(columns= ["customerID"],inplace = True)
print("Final list of required columns :",df.columns.to_list())
```

```python
# Save cleaned data
df.to_csv(CLEANED_DATA_FILE_PATH,index = False)
print("Data is saved successfully")
```
